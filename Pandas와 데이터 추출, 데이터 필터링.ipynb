```python
import pandas as pd

df_dict = dict()

df_dict['이름'] = ['김고려', '최안암', '박정보', '이대학' ]
df_dict['GPA'] = [4.0, 3.9, 4.2, 3.8]
df_dict['소속'] =  ['정보대학', '공과대학', '자유전공학부', '정보대학']
df_dict['출석률'] = ['95%', '98%', '92%', '89%']
df_dict['등록학기'] = [4, 3,7, 2]

df = pd.DataFrame(df_dict)

df
```

```python
cond = df['GPA'] >= 4.0     #df[] : 대괄호가 바로 있으면 열추출 의미
df_new = df.loc[cond]

df_new
```

```python
cond1 = df['소속'] != '공과대학'
cond2 = df['등록학기'] >= 3

cond = cond1 & cond2      #and, or같은 논리연산자들은 bool자료형에만 사용 가능
df_new = df.loc[cond]

df_new
```

```python
cond1 = df['소속'] == '정보대학'
cond2 = df['GPA'] < 4.0

cond = cond1 | cond2      # or은 |
df_new = df.loc[cond]

df_new
```
